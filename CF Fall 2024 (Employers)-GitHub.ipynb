```python
# Employers sorted alphabetically - All details
# Pandas is case-sensitive, so it was putting SOS Care above companies like Santee Cooper
# key = lambda and col: col.str.lower() fixes this
# Also need to fix D.R. Horton being before Daniel Island club
# need to import re
# Function to ignore special characters and lowercase when sorting
def sort_key(employer_name):
    #return re.sub(r'[^A-Za-z0-9 .]+', '', employer_name).lower()
    #return re.sub(r'\.', '', employer_name).lower()
    # Removes punctuation, converts everything to lowercase, and accounts for spaces
    clean_name = re.sub(r'[.,]', '', employer_name).lower().replace(' ', '')
    return clean_name
sorted_df = df.sort_values('Employer', key=lambda col: col.map(sort_key)).reset_index(drop=True)
#sorted_df.shape
#(113,28)
```
